```python
import numpy as np # Â∞éÂÖ• NumPy ÂáΩÂºèÂ∫´ÔºåÁî®ÊñºÊï∏ÂÄºË®àÁÆó
import pandas as pd # Â∞éÂÖ• Pandas ÂáΩÂºèÂ∫´ÔºåÁî®ÊñºË≥áÊñôÂàÜÊûê
import matplotlib.pyplot as plt # Â∞éÂÖ• Matplotlib ÂáΩÂºèÂ∫´ÔºåÁî®ÊñºÁπ™Âúñ
import seaborn as sns # Â∞éÂÖ• Seaborn ÂáΩÂºèÂ∫´ÔºåÁî®ÊñºÁµ±Ë®àË≥áÊñôË¶ñË¶∫Âåñ
import matplotlib.cm as cm # Â∞éÂÖ• Matplotlib ÁöÑÈ°èËâ≤Êò†Â∞ÑÊ®°ÁµÑ
%matplotlib inline

# Áπ™Ë£ΩË≤ìÂí™È†≠ÁöÑËº™ÂªìÔºà‰ΩøÁî®ÂúìÂºßÔºâ
theta = np.linspace(0, np.pi, 200) # ÁîüÊàêÂæû 0 Âà∞ pi ÁöÑ 200 ÂÄãÁ≠âË∑ùÈªû
x_head = np.cos(theta) * 2 # Ë®àÁÆóË≤ìÂí™È†≠ x Â∫ßÊ®ô
y_head = np.sin(theta) * 2 # Ë®àÁÆóË≤ìÂí™È†≠ y Â∫ßÊ®ô

# Áπ™Ë£ΩËÄ≥ÊúµÔºà‰ΩøÁî®‰∫åÊ¨°Ë≤ùËå≤Êõ≤Á∑öÔºâ
t = np.linspace(0, 1, 200) # ÁîüÊàêÂæû 0 Âà∞ 1 ÁöÑ 200 ÂÄãÁ≠âË∑ùÈªû

# Â∑¶ËÄ≥Êúµ
ear1_x = (1 - t) ** 2 * -1.2 + 2 * (1 - t) * t * -1.7 + t ** 2 * -0.8 # Ë®àÁÆóÂ∑¶ËÄ≥ x Â∫ßÊ®ô (Ë≤ùËå≤Êõ≤Á∑ö)
ear1_y = (1 - t) ** 2 * 1.6 + 2 * (1 - t) * t * 2.8 + t ** 2 * 1.8 # Ë®àÁÆóÂ∑¶ËÄ≥ y Â∫ßÊ®ô (Ë≤ùËå≤Êõ≤Á∑ö)

# Âè≥ËÄ≥ÊúµÔºàÂ∞çÁ®±ÁøªËΩâÔºâ
ear2_x = (1 - t) ** 2 * 1.2 + 2 * (1 - t) * t * 1.7 + t ** 2 * 0.8 # Ë®àÁÆóÂè≥ËÄ≥ x Â∫ßÊ®ô (Ë≤ùËå≤Êõ≤Á∑ö)
ear2_y = (1 - t) ** 2 * 1.6 + 2 * (1 - t) * t * 2.8 + t ** 2 * 1.8 # Ë®àÁÆóÂè≥ËÄ≥ y Â∫ßÊ®ô (Ë≤ùËå≤Êõ≤Á∑ö)

# ÂÆöÁæ©ÂΩ©ËôπÊº∏Â±§È°èËâ≤
colors = cm.rainbow(np.linspace(0, 1, len(x_head))) # ‰ΩøÁî®ÂΩ©ËôπÈ°èËâ≤Êò†Â∞ÑÁîüÊàêÈ°èËâ≤ÂàóË°®

# Áπ™Ë£ΩË≤ìÂí™È†≠
fig, ax = plt.subplots(figsize=(6, 6)) # Âª∫Á´ãÁπ™ÂúñÁâ©‰ª∂

# Áπ™Ë£ΩÈ†≠ÈÉ®ÂΩ©ËôπÊº∏Â±§
for i in range(len(x_head) - 1): # Áπ™Ë£ΩÈ†≠ÈÉ®ÔºåÊØèÂÖ©ÂÄãÈªû‰πãÈñìÁï´‰∏ÄÊ¢ùÁ∑ö
    ax.plot(x_head[i:i+2], y_head[i:i+2], color=colors[i], linewidth=6) # Áπ™Ë£ΩÁ∑öÊÆµÔºåË®≠ÂÆöÈ°èËâ≤ÂíåÁ∑öÂØ¨

# Áπ™Ë£ΩËÄ≥ÊúµÂΩ©ËôπÊº∏Â±§
for i in range(len(ear1_x) - 1): # Áπ™Ë£ΩËÄ≥ÊúµÔºåÊØèÂÖ©ÂÄãÈªû‰πãÈñìÁï´‰∏ÄÊ¢ùÁ∑ö
    ax.plot(ear1_x[i:i+2], ear1_y[i:i+2], color=colors[i], linewidth=6) # Áπ™Ë£ΩÂ∑¶ËÄ≥Á∑öÊÆµ
    ax.plot(ear2_x[i:i+2], ear2_y[i:i+2], color=colors[i], linewidth=6) # Áπ™Ë£ΩÂè≥ËÄ≥Á∑öÊÆµ

# Áπ™Ë£ΩÁúºÁùõ
plt.scatter([-1], [1], color='red', label='(-1,1)') # Áπ™Ë£ΩÂ∑¶Áúº
plt.scatter([1], [1], color='red', label='(1,1)') # Áπ™Ë£ΩÂè≥Áúº

# Áπ™Ë£ΩÂò¥Â∑¥
x = np.array([-0.25, 0, 0.25]) # Ë®≠ÂÆöÂò¥Â∑¥ x Â∫ßÊ®ô
y = np.array([1, 0.9, 1]) # Ë®≠ÂÆöÂò¥Â∑¥ y Â∫ßÊ®ô
ax.plot(x, y, color='red', linewidth=3) # Áπ™Ë£ΩÂò¥Â∑¥Êõ≤Á∑ö

# Âä†ÂÖ•ÂùêÊ®ôËª∏
ax.axhline(0, color='black', linewidth=1) # Áπ™Ë£ΩÊ∞¥Âπ≥Á∑ö (x Ëª∏)
ax.axvline(0, color='black', linewidth=1) # Áπ™Ë£ΩÂûÇÁõ¥Á∑ö (y Ëª∏)

# Ë®≠ÂÆöÈ°ØÁ§∫ÁØÑÂúç
ax.set_xlim(-2.5, 2.5) # Ë®≠ÂÆö x Ëª∏ÁØÑÂúç
ax.set_ylim(-2, 3) # Ë®≠ÂÆö y Ëª∏ÁØÑÂúç
ax.set_aspect('equal') # Ë®≠ÂÆö x Âíå y Ëª∏ÊØî‰æãÁõ∏Âêå

# È°ØÁ§∫Â∫ßÊ®ôÊï∏Â≠ó
ax.set_xticks(np.arange(-2.5, 3, 0.5)) # Ë®≠ÂÆö x Ëª∏ÂàªÂ∫¶
ax.set_yticks(np.arange(-2, 4, 0.5)) # Ë®≠ÂÆö y Ëª∏ÂàªÂ∫¶
ax.grid(True, linestyle='--', linewidth=0.5) # Â¢ûÂä†Ê†ºÁ∑ö

# Ê®ôÈ°åÊñáÂ≠ó
ax.set_title('üò∫', fontsize=14) # Ë®≠ÂÆöÂúñË°®Ê®ôÈ°å
```
