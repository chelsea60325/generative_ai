#**畫貓咪圖形**


```python
import numpy as np # 導入 NumPy 函式庫，用於數值計算
import pandas as pd # 導入 Pandas 函式庫，用於資料分析
import matplotlib.pyplot as plt # 導入 Matplotlib 函式庫，用於繪圖
import seaborn as sns # 導入 Seaborn 函式庫，用於統計資料視覺化
import matplotlib.cm as cm # 導入 Matplotlib 的顏色映射模組
%matplotlib inline

# 繪製貓咪頭的輪廓（使用圓弧）
theta = np.linspace(0, np.pi, 200) # 生成從 0 到 pi 的 200 個等距點
x_head = np.cos(theta) * 2 # 計算貓咪頭 x 座標
y_head = np.sin(theta) * 2 # 計算貓咪頭 y 座標

# 繪製耳朵（使用二次貝茲曲線）
t = np.linspace(0, 1, 200) # 生成從 0 到 1 的 200 個等距點

# 左耳朵
ear1_x = (1 - t) ** 2 * -1.2 + 2 * (1 - t) * t * -1.7 + t ** 2 * -0.8 # 計算左耳 x 座標 (貝茲曲線)
ear1_y = (1 - t) ** 2 * 1.6 + 2 * (1 - t) * t * 2.8 + t ** 2 * 1.8 # 計算左耳 y 座標 (貝茲曲線)

# 右耳朵（對稱翻轉）
ear2_x = (1 - t) ** 2 * 1.2 + 2 * (1 - t) * t * 1.7 + t ** 2 * 0.8 # 計算右耳 x 座標 (貝茲曲線)
ear2_y = (1 - t) ** 2 * 1.6 + 2 * (1 - t) * t * 2.8 + t ** 2 * 1.8 # 計算右耳 y 座標 (貝茲曲線)

# 定義彩虹漸層顏色
colors = cm.rainbow(np.linspace(0, 1, len(x_head))) # 使用彩虹顏色映射生成顏色列表

# 繪製貓咪頭
fig, ax = plt.subplots(figsize=(6, 6)) # 建立繪圖物件

# 繪製頭部彩虹漸層
for i in range(len(x_head) - 1): # 繪製頭部，每兩個點之間畫一條線
    ax.plot(x_head[i:i+2], y_head[i:i+2], color=colors[i], linewidth=6) # 繪製線段，設定顏色和線寬

# 繪製耳朵彩虹漸層
for i in range(len(ear1_x) - 1): # 繪製耳朵，每兩個點之間畫一條線
    ax.plot(ear1_x[i:i+2], ear1_y[i:i+2], color=colors[i], linewidth=6) # 繪製左耳線段
    ax.plot(ear2_x[i:i+2], ear2_y[i:i+2], color=colors[i], linewidth=6) # 繪製右耳線段

# 繪製眼睛
plt.scatter([-1], [1], color='red', label='(-1,1)') # 繪製左眼
plt.scatter([1], [1], color='red', label='(1,1)') # 繪製右眼

# 繪製嘴巴
x = np.array([-0.25, 0, 0.25]) # 設定嘴巴 x 座標
y = np.array([1, 0.9, 1]) # 設定嘴巴 y 座標
ax.plot(x, y, color='red', linewidth=3) # 繪製嘴巴曲線

# 加入坐標軸
ax.axhline(0, color='black', linewidth=1) # 繪製水平線 (x 軸)
ax.axvline(0, color='black', linewidth=1) # 繪製垂直線 (y 軸)

# 設定顯示範圍
ax.set_xlim(-2.5, 2.5) # 設定 x 軸範圍
ax.set_ylim(-2, 3) # 設定 y 軸範圍
ax.set_aspect('equal') # 設定 x 和 y 軸比例相同

# 顯示座標數字
ax.set_xticks(np.arange(-2.5, 3, 0.5)) # 設定 x 軸刻度
ax.set_yticks(np.arange(-2, 4, 0.5)) # 設定 y 軸刻度
ax.grid(True, linestyle='--', linewidth=0.5) # 增加格線

# 標題文字
ax.set_title('😺', fontsize=14) # 設定圖表標題
```
